```python
import warnings
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.callbacks import History

# Ignore warnings
warnings.filterwarnings("ignore")

# Initialize the CNN
classifier = Sequential()

# Step 1 - Convolution layer with 64 filters, 3x3 kernel, and ReLU activation
classifier.add(Conv2D(64, (3, 3), input_shape=(64, 64, 3), activation='relu'))

# Step 2 - Pooling layer with 2x2 pool size
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# Adding a second convolution layer with 64 filters and ReLU activation
classifier.add(Conv2D(64, (3, 3), activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# Step 3 - Flattening the pooled feature maps
classifier.add(Flatten())

# Step 4 - Full connection layer with 128 units and ReLU activation
classifier.add(Dense(units=128, activation='relu'))

# Output layer with a single unit and sigmoid activation for binary classification
classifier.add(Dense(units=1, activation='sigmoid'))

# Compile the CNN with Adam optimizer and binary cross-entropy loss
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Define the data augmentation for the training set
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)

# Define the data augmentation for the testing set
test_datagen = ImageDataGenerator(rescale=1./255)

# Define the training set
training_set = train_datagen.flow_from_directory(
    'training_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')

# Define the testing set
test_set = test_datagen.flow_from_directory(
    'testing_set',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')

# Train the model and save training history for plotting
history = classifier.fit(
    training_set,
    steps_per_epoch=5000,
    epochs=10,
    validation_data=test_set,
    validation_steps=1000)

# Plotting training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plotting training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Verifying the model by giving samples of cell to detect malaria
test_image = load_img('single_prediction/healthy.JPG', target_size=(64, 64))
test_image = img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
result = classifier.predict(test_image)

# Print the prediction
if result[0][0] == 1:
    prediction = 'Healthy'
else:
    prediction = 'UnHealthy'
print("The sample is predicted as:", prediction)

# Evaluate the model on the test set to get the confusion matrix
Y_pred = classifier.predict(test_set)
y_pred = (Y_pred > 0.5).astype(int)
y_true = test_set.classes

# Generate confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=test_set.class_indices)
disp.plot(cmap=plt.cm.Blues)
plt.show()

# Placeholder for comparing the results with state-of-the-art models
# Assuming `state_of_the_art_accuracies` is a list of accuracies from other models
state_of_the_art_accuracies = [0.95, 0.92, 0.89]  # Example accuracies
model_accuracy = history.history['val_accuracy'][-1]

plt.figure()
plt.bar(['Our Model'] + ['Model ' + str(i) for i in range(1, len(state_of_the_art_accuracies) + 1)], 
        [model_accuracy] + state_of_the_art_accuracies)
plt.ylabel('Accuracy')
plt.title('Comparison of Model Accuracies')
plt.show()
```
